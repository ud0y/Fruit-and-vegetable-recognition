```python
# Importing all necessary libraries
import cv2
import os

def video2image(fileName, folderName):
    '''
    param: filename - string, path of that file as string
    '''

    # Read the video from specified path
    cam = cv2.VideoCapture(fileName)

    try:
        # creating a folder named data
        if not os.path.exists(folderName):
            os.makedirs(folderName)

    # if not created then raise error
    except OSError:
        print ('Error: Creating directory of data')

    # frame
    currentframe = 0

    while(True):
        # reading from frame
        ret,frame = cam.read()

        if ret:
            # if video is still left continue creating images
            name = './'+ folderName +'/frame' + str(currentframe) + '.jpg'
            print ('Creating...' + name)

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 1

        else:
            break

    # Release all space and windows once done
    cam.release()
    cv2.destroyAllWindows()
    
    return currentframe
```

```python
from removebg import RemoveBg
import os

def removeBackground(fileName, folderName):
    rmbg = RemoveBg("PPpu7i6DjiTaBZHaxd5wq2D7", "error.log")
    try:
        # creating a folder named data
        if not os.path.exists('r'+folderName):
            os.makedirs('r'+folderName)

    # if not created then raise error
    except OSError:
        print ('Error: Creating directory of data')
        
    rmbg.remove_background_from_img_file(fileName, 
                                         size='small', 
                                         bg="fff", 
                                         bg_type='color', 
                                         position='center', 
                                         crop='8px',
                                         new_file_name="./r" + folderName + '/' + fileName)
```

```python
from PIL import Image

def ImageResize(fileName, folderName, cnt=0):
    #read the image
    im = Image.open(fileName)

    #image size
    size = (100,100)
    box = (100,100,100,100)
    #resize image
    out = im.resize(size, box=box)
    #save resized image
    out.save('./resized/'+folderName+'/image'+str(cnt)+'.jpg')
    cnt += 1
    return cnt
```

```python
import os

def preprocess(filepath):
    videos = os.listdir(filepath)
    
    for video in videos:
        folderName = video.split('_')[0]
        
        print('Video: ' + video + ' Folder: ' + folderName)
        
        cf = video2image(fileName=video, folderName=folderName)
        
        images = os.listdir(folderName)
        
        print(len(images))
        
        for image in images:
            removeBackground(image, folderName)
        
        removedBackgroungImages = os.listdir('./r'+folderName)
        cnt = 0
        
        for image in removedBackgroungImages:
            cnt = ImageResize(image, folderName, cnt)
            
        print(f"Process for {folderName} done...")
        
    print("All done..")
```

```python
preprocess('dataset')
```
