```python
# Importing all necessary libraries
import cv2
import os

def video2image(filePath, folderName, countFrame):
    '''
    param: filePath - string, path of that file as string
    param: folderName -  string, output folder name
    param: countFrame - int, initial frame number
    '''

    # Read the video from specified path
    cam = cv2.VideoCapture(filePath)

    try:
        # creating a folder named data
        if not os.path.exists(folderName):
            os.makedirs(folderName)

    # if not created then raise error
    except OSError:
        print ('Error: Creating directory of data')

    # frame
    currentframe = countFrame

    while(True):
        # reading from frame
        ret,frame = cam.read()

        if ret:
            # if video is still left continue creating images
            name = './'+ folderName +'/frame' + str(currentframe) + '.jpg'
            print ('Creating...' + name)

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 1

        else:
            break

    # Release all space and windows once done
    cam.release()
    cv2.destroyAllWindows()
    return countFrame
```

```python
import os

videos = os.listdir('./Raw Video Dataset')
videos
```

```python
cf = 0

for video in videos:
    folderName = video.split('.')[0]
    print(folderName)
    
    cf = video2image(filePath='./Raw Video Dataset/'+video, folderName='./rawTest/'+folderName, countFrame=cf)
```

```python
from removebg import RemoveBg
import os

def removeBackground(fileName, folderName, path):
    rmbg = RemoveBg("PPpu7i6DjiTaBZHaxd5wq2D7", "error.log")
    try:
        # creating a folder named data
        if not os.path.exists('./removedBG/'+folderName):
            os.makedirs('./removedBG/'+folderName)

    # if not created then raise error
    except OSError:
        print ('Error: Creating directory of data')
        
    rmbg.remove_background_from_img_file(img_file_path = path + '/' + fileName,
                                         size='small', 
                                         bg="fff", 
                                         bg_type='color', 
                                         position='center', 
                                         crop='8px',
                                         new_file_name="./removedBG/" + folderName + '/' + fileName)
```

```python
from PIL import Image

def ImageResize(filePath, folderName, cnt):
    #read the image
    im = Image.open(filePath)

    #image size
    size = (100,100)
    box = (100,100,100,100)
    #resize image
    out = im.resize(size, box=box)
    #save resized image
    out.save('./resized/'+folderName+'/image'+str(cnt)+'.jpg')
    cnt += 1
    return cnt
```

```python
import os

def preprocess(path, folderName):
    images = os.listdir(path)
    cnt = 0
    
    for image in images:
        removeBackground(fileName=image, folderName=folderName, path=path)
        cnt += 1

    removedBackgroungImages = os.listdir('./removedBG/'+folderName)
    cnt = 0

    for image in removedBackgroungImages:
        cnt = ImageResize('./removedBG/'+image, folderName, cnt)

    prevFolder = folderName
    print(f"Process for {folderName} done...")
        
    print("All done..")
```
